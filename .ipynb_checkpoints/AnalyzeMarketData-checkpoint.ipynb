```python
import pandas as pd 
import numpy as np 
import chardet
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from matplotlib import style
from pytrends.request import TrendReq
```

```python
#import real estate market data, this can't be stored on Github because the file is too large
core_metrics_df = pd.read_csv('/home/user0/Desktop/RealEstateProject/RealEstateTimeSeries.csv', dtype={'ZipCode': "category"})
#set index to zipcode 
core_metrics_df.set_index('ZipCode', inplace=True)
print(core_metrics_df.head())
```

```python
#search a zipcode
zip = '10516' #Cold Spring 
print(core_metrics_df.loc[zip].head())
```

```python
#create a new dataframe that contains timeseries pricing data and a single feature 
main_df = pd.DataFrame({'Month': core_metrics_df.loc[zip]['Month'],
                        'Median Listing Price': core_metrics_df.loc[zip]['Median Listing Price'],
                        'Feature': core_metrics_df.loc[zip]['Active Listing Count Y/Y']})
main_df.drop(main_df.index)
main_df.set_index('Month', inplace=True)
main_df.index = pd.to_datetime(main_df.index)
main_df.sort_values(by='Month', inplace=True)
print(main_df.tail())
```

```python
#use pytrend to get google trends data on specific searches (in this case searching for "some location + real estate")
pytrend = TrendReq()
```

```python
search = 'Cold Spring NY homes'
pytrend.build_payload(kw_list=[search], timeframe='all')
```

```python
interest_over_time_df = pytrend.interest_over_time()
interest_over_time_df.drop(interest_over_time_df.index[-1], inplace=True)
interest_over_time_df['Average Interest Over Time'] = interest_over_time_df[search].rolling(window=12).mean()
print(interest_over_time_df.head())
```

```python
ax1 = plt.gca()
interest_over_time_df.plot(kind='line',y='Average Interest Over Time',ax=ax1)
```

```python
main_df = pd.concat([main_df, interest_over_time_df['Average Interest Over Time']], axis=1)
```

```python
print(main_df.head())
```

```python
def graph_data(val1,val2): 
    style.use('ggplot')
    fig, ax1 = plt.subplots()
    ax1.plot(val1.index, val1)
    ax2 = ax1.twinx()
    ax2.plot(val2, c='g')
    ax1.set_xlabel('Year')
    ax1.set_ylabel(val1.name)
    ax2.set_ylabel(val2.name)
    ax1.legend(loc=0, fontsize=10, frameon=True).get_frame().set_edgecolor('blue')
    ax2.legend(loc=1, fontsize=10, frameon=True).get_frame().set_edgecolor('blue')
    
graph_data(main_df['Median Listing Price'], main_df['Average Interest Over Time']['2012-01-01':])
```
